<h5>Dataset and Inputs</h5>

<p>The dataset presented here (the ELLIPSE corpus) comprises argumentative essays written by
8th-12th grade English Language Learners (ELLs). The essays have been scored according to six
analytic measures: cohesion, syntax, vocabulary, phraseology, grammar, and conventions. Each
measure represents a component of proficiency in essay writing, with greater scores
corresponding to greater proficiency in that measure. The scores range from 1.0 to 5.0 in
increments of 0.5. The goal is to predict the score of each of the six measures for the essays
given in the test set.</p>

Bringing real-world knowledge into this problem, I want to observe how the length of the text or the number of stop words can possibly impact the scores

<h2>Section 3: Data Pre Processing</h2>

<p> In data pre-processing, let us tokenize and lower the text case of the full text so that the data can be vectorized for algorithms to consume. Stop words and punctuations shouldn't be eliminated since we are evaluating essay. However, stopwords will be considered separately during vectorization</p>

```python
train['tokens'] = train.full_text.apply(lambda x: nltk.word_tokenize(x.lower()))
train['text_len'] = train.tokens.apply(lambda x: len(x))
train['stop_words_len'] = train.tokens.apply(lambda x: len([w for w in x if w in eng_stop_words]))
train.head()
```

```python
# calculate the correlation matrix
corr = train[ATTRS + ['stop_words_len', 'text_len']].corr()

# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

<p> Now we have an idea about the data and we can move towards implemntation of algorithms. For instance, Metrics and stop words len are inversely correlated. So, if one uses more words and more stop words then it is likely that a person is going to get less scores in all the metrics</p>

# Implementation
After reading and analysing the data, Let's start simple with count vectorization for feature extraction. Then we will test the data on regression models. We'll see if alternate techniques provide refinement. Once the results are established, we will go with the best model.

<h4>Defining Metric</h4>

```python

def RMSE(y, y_hat):
    return np.sqrt(mean_squared_error(y,y_hat))

def MCRMSE(y_df, y_hat_df):
    size = y_df.shape[1]
    mcrmse = 0
    for i in range(size):
        mcrmse+=RMSE(y_df[:, i], y_hat_df[:, i])
    return mcrmse/size
```

```python
X = train['full_text'].values
y = train[ATTRS].values
```

```python
X.shape
```

```python
y.shape
```

<p>Let us use CountVectorizer to vectorize student essays</p>

```python
vectorizer = CountVectorizer()
vectorizer.fit(X)
vector = vectorizer.transform(X)
```

```python
X = vector.astype(np.float64)
```

<h5>Splitting our data for training and testing</h5>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

<h2>Candidate Model Exploration</h2>

<h3>Linear Regression</h3>

```python
params_reg = {
    "n_estimators": 1000,
    "verbose": -1
}
```

```python
model = MultiOutputRegressor(LinearRegression())
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(MCRMSE(y_test,y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
y_pred = model.predict(X_test)
print(MCRMSE(y_test,y_pred))
print(mean_absolute_error(y_test, y_pred))
```
